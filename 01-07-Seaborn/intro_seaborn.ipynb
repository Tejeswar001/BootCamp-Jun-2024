## Line Plot

The basic parameters of `sns.lineplot()` are:

- `x`, `y`: These specify the variables to plot on the x and y axes, respectively.
- `data`: This is the dataset containing the variables you want to plot.
- `hue`: This lets you group the data by a third categorical variable.
- `style`: This lets you further differentiate groups by applying different line styles.
- `markers`: This allows you to add markers to the lines for better visibility of data points.
